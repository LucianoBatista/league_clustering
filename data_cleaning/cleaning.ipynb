### Agora todos estão com mesmo número de colunas, podemos concatenar

### Apesar de possuirmos muitas observações, muitos dados vieram com erros, consequência do limite de requests na API da RIOT. Então precisamos tratar esses erros

```python
df_final['game_ids'].value_counts()
```

Além dos erros espalhados pelo dataset que precisamos tratar, vemos que existem game_ids repetidos, ou seja, a mesma partida foi contabilizada mais de uma vez. Dessa forma, também será preciso tratar as observações duplicadas.

Para os erros observados, vamos substituir por _np.nan_, para auxiliar no tratamento posterior.

```python
df_final = df_final.replace('error', np.nan)
df_final = df_final.replace('erro', np.nan)
```

```python
df_final.head()
```

O drop dos valores NaN será feito inicialmente em relação à coluna _game_ids_. Isso por que, se nosso algoritmo falhou em encontrar um game_id, consequentemente ele falhou em encontrar todo o resto das estatísticas.

```python
df_final = df_final.dropna(subset=['game_ids'])
```

```python
# identificação das duplicatas
df_final['game_ids'].value_counts()
```

```python
# removendo duplicadas em game_ids
df_final = df_final.drop_duplicates(subset=['game_ids'])
```

Também verificou-se que a maioria dos dados que não possuíam dados na coluna _gameDuration_ também apresentavam um alto valor de valores missing. Então decidimos em dropar algumas linhas em relação a essa coluna também.

```python
df_final = df_final.dropna(subset=['gameDuration'])
```
